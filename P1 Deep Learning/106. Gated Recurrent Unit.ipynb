        <p>GRU (Gated Recurrent Unit) is a recurrent neural network that slightly improves LSTM, proposed by Cho et al. (2014). It combines the forget gate and input gate into a single "update gate", while merging the cell state and hidden state, and making other changes. The GRU model is simpler compared to the standard LSTM model and is increasingly receiving widespread attention and application. The main difference between GRU and LSTM is the design of the gating mechanism. GRU only uses one update gate, which controls the flow of information and updates the state, while reducing the number of parameters and computation. Compared to LSTM, the GRU model is more concise, easier to train, and has achieved quite good performance in some tasks.

## GRU
Before building the model, let's review the parameters of the new layer:

**nn.GRU** is a gated recurrent unit (GRU) module implemented in PyTorch, used to construct neural network layers with GRU units. The common parameters are as follows:

`input_size`: The size of the input features.

`hidden_size`: The number of features in the hidden state.

`num_layers`: (optional) The number of GRU layers. Default is 1.

`bias`: (optional) Whether to include bias terms. Default is True.

`batch_first`: (optional) If True, input and output tensors are provided as (batch, seq, feature). If False, input and output tensors are provided as (seq, batch, feature). Default is False.

`dropout`: (optional) If non-zero, applies dropout to the GRU output. Default is 0.

`bidirectional`: (optional) If True, creates a bidirectional GRU. Default is False.

```python
import torch
import torch.nn as nn

class GRU(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(GRU, self).__init__()
        self.lstm = nn.GRU(input_size = input_size, 
                            num_layers = 2,
                            hidden_size = hidden_size,  
                            batch_first = True)
        
        self.fc = nn.Linear(hidden_size, output_size) 
        
    def forward(self, x):  
        out, _ = self.lstm(x)
        out = self.fc(out[:, -1, :])
        return out
```
